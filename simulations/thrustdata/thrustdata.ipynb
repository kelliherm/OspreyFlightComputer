### Model Rocket Engine Thrust Data

Based on information from [thrustcurve.org](https://www.thrustcurve.org/), using data uploaded by contributors Mike Haberer and John Coker. Without them, this would not otherwise be possible.

```python
import matplotlib.pyplot as plt
import csv
```

```python
def GetData(file, type):
    filelocation = file

    time = []
    thrust = []

    with open(filelocation, mode ='r') as file:
        csvFile = csv.reader(file)

        next(csvFile)

        for line in csvFile:         
            time.append(float(line[0]))
            thrust.append(float(line[1]))
        
    if type == 'time':
        return time
    elif type == 'thrust':
        return thrust
```

```python
plt.plot(GetData('Estes_E16.csv', 'time'), GetData('Estes_E16.csv', 'thrust'), 'r-')
plt.title('Estes E16 Motor')
plt.xlabel('Time (s)')
plt.ylabel('Thrust (N)')
plt.show()
```
